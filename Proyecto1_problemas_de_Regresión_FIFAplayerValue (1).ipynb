## Outliers

### Experimento 1. Bosques de Aislamiento

deteccion de ouliers con multidimensiones a partir del metodo de Isolation Forest.

```python
df1 = df[['Valor_en_euros','Salario_en_euros']]
mi_array = np.array([df1['Valor_en_euros'],df1['Salario_en_euros']])

from sklearn.ensemble import IsolationForest

# Crear un array con dos columnas: 'value_euro' y 'age'
mi_array = df1[['Salario_en_euros','Valor_en_euros']].values
resultados = np.zeros((3, mi_array.shape[0]))

# Bosques de Aislamiento con diferente contaminación
c = [0.000001, 0.05, 0.1]
for i in range(len(c)):
    modelo = IsolationForest(contamination=c[i]).fit(mi_array)
    resultados[i] = modelo.predict(mi_array)

# Graficar datos anómalos
plt.set_cmap("jet")
fig = plt.figure(figsize=(13, 4))

for i in range(len(c)):
    ax = fig.add_subplot(1, 3, i+1)
    ax.scatter(mi_array[resultados[i]==-1][:, 0],
               mi_array[resultados[i]==-1][:, 1],
               c="skyblue", marker="s", s=500)
    ax.scatter(mi_array[:, 0],
               mi_array[:, 1],
               c=range(df1.size//2), marker="x",
               s=500, alpha=0.6)
    ax.set_title("Contaminación: %0.2f" % c[i], size=18, color="purple")
    ax.set_ylabel("Salario_en_euros", size=10)
    ax.set_xlabel("Valor Venta", size=14)

plt.show()
```

```python
# Identificar los índices de los outliers detectados con contaminación de 0.05
outliers_indices = np.where(resultados[1] == -1)[0]

# Eliminar los outliers del DataFrame original
dfBOSQUE = df.drop(outliers_indices)
```

# Regresión lineal simple

Para predecir la variable "Valor en euros", construiremos dos regresiones lineales con distintas x. La primera será con "Calificación general", la segunda con "Salario en euros".

### Reg.simple. X="Calificación General"

```python
from sklearn.model_selection import train_test_split
```

```python
y = df["Valor_en_euros"]
X = df[["Calificación_general"]]
```

```python
# Dividimos los datos en train y test
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
print("Forma de X_train:", X_train.shape)
print("Forma de X_test:", X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Crear un modelo de regresión lineal
model = LinearRegression()

# Ajustar el modelo a los datos de entrenamiento
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular el coeficiente de determinación (R^2) para evaluar el rendimiento del modelo
r_squared = model.score(X_test, y_test)

#calculamos el MSE
mse_clasica = mean_squared_error(y_test, y_pred)

# Visualizar la regresión lineal
plt.scatter(X_test, y_test, color='blue', label='Test')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Regresión lineal')
plt.xlabel('Calificación General')
plt.ylabel('Valor en Euros')
plt.legend()
plt.show()


# Imprimir el coeficiente de determinación (R^2) y MSE
print("Coeficiente de determinación (R^2):", r_squared)
print("Mean Square Error:", mse_clasica)
```

### modelo con la regresión Ridge haciendo Cross Validation:

# Regresión K-Nearest-Neighbor

```python
from matplotlib.colors import ListedColormap
import matplotlib.patches as mpatches
import seaborn as sb

%matplotlib inline
plt.rcParams['figure.figsize'] = (16, 9)
plt.style.use('ggplot')

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

```python
X = df.drop('Valor_en_euros', axis=1)
y = df["Valor_en_euros"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print("Forma de X_train:", X_train.shape)
print("Forma de X_test:", X_test.shape)
```

Buscamos el mejor valor de K haciendo CrossValidation

```python
for k in range(2, 11, 1):

    # Creamos el modelo para un dado valor de i
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(X_train, y_train)

    # Calculamos los scores al hacer cross-validation con 5 folds
    scores = cross_val_score(knn, X_train, y_train, cv = 5)

    # Imprimimos el valor de alpha y el score medio
    print(f'n_neighbors {k} - Score promedio de 5 folds = {scores.mean()}')
```

El mejor valor de K es 3:

```python
knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
pred_y = knn.predict(X_test)
```

```python
# Calcular el coeficiente de determinación (R^2)
r2 = r2_score(y_test, pred_y)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, pred_y)

print("Coeficiente de determinación (R^2):",r2)
print("Mean Square Error:",mse)
```
